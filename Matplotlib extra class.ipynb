# Matplotlib tutorial for NPTEL

# Scatter plot

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
```

```python
x=[1,2,3,4,5,6]
bumrah=[140,136,145,150,148,125]
jadeja=[80,90,87,100,88,96]
shami=[120,135,132,140,142,136]
```

```python
plt.plot(x,bumrah,color='black',linestyle='-',marker='+',markeredgecolor='red',markersize=6, linewidth=2, alpha=1,label='Bumrah' )
plt.plot(x,jadeja,color='blue',linestyle='-',marker='+',markeredgecolor='yellow',markersize=6, linewidth=2, alpha=1,label='Jadeja' )
plt.plot(x,shami,color='red',linestyle='-',marker='+',markeredgecolor='green',markersize=6, linewidth=2, alpha=1,label='Shami' )
plt.legend(fontsize=10,loc='upper left')
plt.ylabel('Ball speed',fontsize=12)
plt.xlabel('Ball no',fontsize=12)
plt.title('Bowling speed in an over',color='r',fontsize='15')
```

# Barplot

```python
batsmen=['Rohit',"Virat",'Subhman', 'Hardik', 'Rahul']
xaxis=np.arange(len(batsmen))
scores_pak=[70,78,120,56,90]
scores_sa=[80,53,45,65,65]
```

```python
plt.xticks(xaxis,batsmen)
plt.bar(xaxis-0.2,scores_pak,width=0.4,label='Scores against \nPAK')
plt.bar(xaxis+0.2,scores_sa,width=0.4,label='Scores against \nSA')
plt.legend()
```

# Histograms

```python
grades=[70,67,56,78,45,78,98,76,99,45,32,67,45,87,66,75,87,86,65,87,56,75,45,15,24,56,67,87,65,86,80,84,65,65,87]
plt.hist(grades,bins=np.arange(0,110,10),rwidth=0.95)
plt.xticks(np.arange(0,110,10))
plt.yticks(np.arange(0,11,1))
plt.grid(alpha=0.1)
```

# Subplots and saving figures

```python
x=np.arange(1,2000,1)
y1=x**2
y2=np.log10(x)
```

```python
fig,ax=plt.subplots(2,2,sharey=True,figsize=(12,12))
ax[0,0].set_facecolor('b')
ax[1,0].set_facecolor('r')
ax[0,0].plot(x,y1, color='black')
ax[0,0].set_xlabel('square of X')
ax[1,0].plot(x,y2,color='black')
ax[1,0].set_xlabel('Natural log of X')
ax[1,1].grid()
plt.suptitle('Using subplots')
plt.savefig('example.pdf',dpi=300)
```

# using tickers in subplots

```python
x=[1,2,3,4,5,6]
bumrah=[140,136,145,150,148,125]
jadeja=[80,90,87,100,88,96]
shami=[120,135,132,140,142,136]
```

```python
fig,ax=plt.subplots(1,1)
ax.plot(x,bumrah,color='black',linestyle='-',marker='+',markeredgecolor='red',markersize=6, linewidth=2, alpha=1,label='Bumrah' )
ax.plot(x,jadeja,color='blue',linestyle='-',marker='+',markeredgecolor='yellow',markersize=6, linewidth=2, alpha=1,label='Jadeja' )
ax.plot(x,shami,color='red',linestyle='-',marker='+',markeredgecolor='green',markersize=6, linewidth=2, alpha=1,label='Shami' )
ax.legend(fontsize=10,loc='upper left')
ax.set_ylabel('Ball speed',fontsize=12)
ax.set_xlabel('Ball no',fontsize=12)
ax.set_title('Bowling speed in an over',color='r',fontsize='15')
ax.xaxis.set_major_locator(ticker.FixedLocator([1,3,6])) 
```
